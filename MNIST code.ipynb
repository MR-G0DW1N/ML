```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
cifar = tf.keras.datasets.cifar100
```

```python
(x_train, y_train), (x_test, y_test) = cifar.load_data()
```

```python
(mxtr, mytr), (mxte, myte) = tf.keras.datasets.mnist.load_data()
```

```python
mxtr = mxtr/255.0
mxte = mxte/255.0

    
```

```python
mxtr.shape
```

```python
num_images = 5

# Generate random indices
random_indices = np.random.randint(0, len(x_test), size=num_images)

# Extract chosen images and labels
chosen_images = mxtr[random_indices]
chosen_labels = mytr[random_indices]
```

```python
for i in range(num_images):
    plt.imshow(chosen_images[i])
    plt.title(f"Label: {chosen_labels[i]}")
    plt.show()
    
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10)
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.fit(mxtr, mytr, epochs=10)
```

```python
test_loss, test_acc = model.evaluate(mxte,  myte, verbose=2)

print('\nTest accuracy:', test_acc)
```
